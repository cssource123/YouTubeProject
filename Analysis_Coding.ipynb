```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "C:\\Users\\Ben Huang\\Desktop\\Youtube_Analysis\\USvideos.csv"
file_2 = "C:\\Users\\Ben Huang\\Desktop\\Youtube_Analysis\\US_category_id.csv"
```

```python
USVideos = pd.read_csv(file)
US_Id = pd.read_csv(file_2)
```

```python
USVideos.head()
```

```python
US_Id.head()
```

```python
US_merged = pd.merge(USVideos, US_Id, on=['category_id'])
US_merged.head()
```

```python
US_merged_fix = US_merged.rename(index=str, columns={'items__snippet__title':'category_type'})
```

```python
US_merged_fix.head()
```

```python
US_merged_categorized = US_merged_fix.groupby(['category_type']).sum()
US_merged_categorized_True = US_merged_categorized.drop(['Entertainment'], axis=0)
US_merged_categorized_True.head()
```

```python
US_views_category = US_merged_categorized_True[['views']]
US_views_category.head()
```

```python
US_views_chart = US_views_category.plot.bar(figsize=(20,10))
US_views_chart.set_ylim(0,45000000000)
US_views_chart.set_ylabel('Views To The 10 Billions')
plt.savefig('US_Category_Views.png',dpi=1000)
```

```python
US_total_videos = US_merged_fix.groupby(['category_type']).count()
US_total_videos_true = US_total_videos.drop(['Entertainment'],axis=0)
US_total_videos_true.head()
```

```python
US_total_videos_final = US_total_videos_true[['video_id']]
US_total_videos_final.head()
```

```python
US_total_videos_graph = US_total_videos_final.plot.bar(figsize=(20,10))
US_total_videos_graph.set_ylabel('Number of Videos Made')
plt.savefig('US_Total_Videos.png',dpi=1000)
```
